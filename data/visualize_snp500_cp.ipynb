```python
import matplotlib.pyplot as plt
from data.snp500 import snp500_daily_cp, snp500_weekly_cp, snp500_monthly_cp

# Plot daily closing prices
plt.figure(figsize=(12, 6))
plt.plot(snp500_daily_cp.index, snp500_daily_cp['Close'], label='Daily Closing Price', color='blue')
plt.title("S&P 500 Daily Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot weekly closing prices
plt.figure(figsize=(12, 6))
plt.plot(snp500_weekly_cp.index, snp500_weekly_cp['Close'], label='Weekly Closing Price', color='orange')
plt.title("S&P 500 Weekly Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot monthly closing prices
plt.figure(figsize=(12, 6))
plt.plot(snp500_monthly_cp.index, snp500_monthly_cp['Close'], label='Monthly Closing Price', color='green')
plt.title("S&P 500 Monthly Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(snp500_daily_cp.index, snp500_daily_cp['Close'], label='Daily', alpha=0.5)
plt.plot(snp500_weekly_cp.index, snp500_weekly_cp['Close'], label='Weekly', alpha=0.7)
plt.plot(snp500_monthly_cp.index, snp500_monthly_cp['Close'], label='Monthly', alpha=0.9)
plt.title("S&P 500 Closing Prices Across Frequencies")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
